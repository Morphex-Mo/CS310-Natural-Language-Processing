## CS310 Natural Language Processing
## Lab 13: Coding Agent

In this lab, we will play with the [Mini-Coding-Agent](https://github.com/rasbt/mini-coding-agent).

1. You need to use ollama to serve an LLM locally, for instance, `qwen3.5-4b`.

2. Follow the instructions (https://github.com/rasbt/mini-coding-agent) to run the agent.

3. See if you can implement a simple program (e.g., quicksort) and test it by talking to the agent.
